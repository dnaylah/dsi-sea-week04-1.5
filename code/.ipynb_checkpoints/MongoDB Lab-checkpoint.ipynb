## Navigate to wikipedia and pull down a table of your choice. Here is a sample of what I did to pull down the Harrison Ford movies table to help you out.

```python
import urllib2
from bs4 import BeautifulSoup
import json
import pandas as pd

soup = BeautifulSoup(urllib2.urlopen('https://en.wikipedia.org/wiki/List_of_tallest_buildings', "lxml").read())
tables = soup.find_all('table', class_="wikitable")
listy = []
for idx, table in enumerate(tables):
    if idx == 1:
        for idx2, row in enumerate(table.find_all('tr')):
            datay = row.find_all('td')
            if len(datay) > 3:
                listy += [{'Rank':datay[0].text,'Name':datay[-7].text,'City':datay[-6].text, 'Country':datay[-5].text,
                         'Height in feet':datay[-4].text,'Height in meters':datay[-3].text, 'Floors':datay[-2].text}]
            #print row
tallest = pd.DataFrame(listy)
print tallest.head()

del tallest['Height in meters']
del tallest['Rank']
```
